```python
import pyodbc
import pandas as pd
import os

# Database connection parameters
server = 'IbrahimHegaziVM'
database = 'DWH'

# Output directory
output_path = r'C:\Users\noobh\Desktop\AI_Assistant_Usage_in_Student_Life_Full_On_Prem_ETL_Pipeline\data\cleaned and modeled'

# Create output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)

# Using Windows Authentication
conn_str = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes'

try:
    # Establish connection
    conn = pyodbc.connect(conn_str)
    print("✅ Connected successfully!")
    
    # List of 7 tables to extract
    tables = [
        'DB.DimDate',
        'DB.DimAIAssistanceLevel',
        'DB.DimDiscipline',
        'DB.DimOutcome',
        'DB.DimStudentLevel',
        'DB.DimTaskType',
        'DB.FactSession'
    ]
    
    # Dictionary to store DataFrames
    dataframes = {}
    
    # Extract each table
    print("\n🔄 Extracting tables...")
    for table in tables:
        query = f"SELECT * FROM {table}"
        df = pd.read_sql(query, conn)
        dataframes[table] = df
        
        # Get table name without schema
        table_name = table.split('.')[-1]
        
        # Full path for CSV file
        csv_path = os.path.join(output_path, f'{table_name}.csv')
        
        # Save to CSV in the specified folder
        df.to_csv(csv_path, index=False)
        print(f"✅ {table}: {len(df):,} rows, {len(df.columns)} columns → saved to {csv_path}")
    
    # Close connection
    conn.close()
    print(f"\n✨ All tables extracted successfully!")
    print(f"📁 Files saved to: {output_path}")
    
    # Display summary of extracted data
    print("\n📊 Extraction Summary:")
    for table, df in dataframes.items():
        print(f"  • {table}: {df.shape[0]} rows × {df.shape[1]} columns")
    
except Exception as e:
    print(f"❌ Error: {e}")
```
